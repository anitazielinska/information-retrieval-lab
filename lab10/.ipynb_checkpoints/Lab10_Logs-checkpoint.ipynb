```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

1) Log file is loaded

```python
file = open("log.txt", "r") 
lines = file.readlines()
for l in range(3): print(lines[l][:-1])
```

2) Cass "Request" parses and keeps data of a single requests. Complete the TODO, where a given "line" (string) should be parsed in order to dervie the ip of the host making the request, time of the request (in seconds), and target page (name).

```python
class Request:
    def __init__(self, line):
        line = line.split(" ")
        
        self.ip = line[0]
        # convert hours + minutes + seconds -> total number of secondsw
        self.time = int(line[1].split(":")[-1]) + int(line[1].split(":")[-2])*60 + int(line[1].split(":")[-3])*3600
        self.page = line[4]
        
    def __repr__(self):
        return self.ip + "     " +  str(datetime.timedelta(seconds=self.time)) + " -> " + self.page
```

3) Each line of log.txt is parsed and the Request objects are added to the list

```python
reqs = [Request(line) for line in lines]
for l in range(10): print(reqs[l]) 
```

# Exercise

Your task is to analyze the log file. You should indentify the users and sessions. Then, compute the following: <br>
3.1) the number of sessions (should be 1000 :) ), <br>
3.2) the number of users, <br>

3.6) use the distribution plot to visualize sessions' starting times (hours). <br>
3.7) which web page is the most common entry point?; for each page plot the number of times it has been an entry page,  <br>
3.8) which web page is the most common departure point?; for each page plot the number of times the user stopped a session while being on this page, <br>
3.9) for each page plot the average time the user stayed on this page <br>
3.10) print the fist 10 sessions (paths). <br>

You may find the following classes helpful and data.

```python
class Session:
    def __init__(self):
        self.startTime = 0
        self.stopTime = 0
        self.requests = []
        
    def __repr__(self):
        print("SESSION")
        print(self.startTime)
        print(self.stopTime)
        STR = ""
        for r in self.requests:
            print(r)
            STR += ("  " + r.page + "    " + str(datetime.timedelta(seconds=r.time)) + "\n")
        return STR

class Computer:
    def __init__(self):
        self.ip = ""
        self.sessions = []
    def __repr__(self):
        return self.ip
    
    
PAGES = ["Shop.html", "AddProduct.html", "Summary.html", "Payment.html","About.html", "Contact.html"]

LINKS = [[1, 4],
         [2,4],
         [3],
         [],
         [1,5],
         []]
    
computers = {}
for req in reqs:
    # New Computer
    if req.ip not in computers:
        computers[req.ip] = Computer()
        computers[req.ip].ip = req.ip
        sess = Session()
        sess.startTime = req.time
        sess.stopTime = req.time + 900
        sess.requests.append(req)
        computers[req.ip].sessions.append(sess)
    # Existing
    else:
        # If more time has passed than is acceptable or there is no link to this page from last then create new session
        if computers[req.ip].sessions[-1].stopTime < req.time\
        or PAGES.index(req.page) not in LINKS[PAGES.index(computers[req.ip].sessions[-1].requests[-1].page)]:
            sess = Session()
            sess.startTime = req.time
            sess.stopTime = req.time + 900
            sess.requests.append(req)
            computers[req.ip].sessions.append(sess)
        # Refresh time window and add request to current Session
        else:
            computers[req.ip].sessions[-1].stopTime = req.time + 900
            computers[req.ip].sessions[-1].requests.append(req)
        
    
# --- 1 ---
session_count = 0
for ip, comp in computers.items():
    session_count += len(comp.sessions)
print("1.", session_count)

# --- 2 ---
user_count = len(computers.keys())
print("2.", user_count)

# --- 3 ---
avg_sessions = session_count/user_count
print("3.", round(avg_sessions, 3))

# --- 4 ---
session_times = []
for i in computers.values():
    for s in i.sessions:
        session_times.append((s.stopTime - s.startTime)/60)

plt.xlabel("długość sesji(min)")
plt.ylabel("prawdopodobieństwo")
plt.title("--- 4 ---")
count, bins, ignored = plt.hist(session_times, 50, alpha=0.5, density=True)
plt.show()

# --- 5 ---
session_lengths = []
for i in computers.values():
    for s in i.sessions:
        session_lengths.append(len(s.requests))

plt.xlabel("długość sesji(liczba żądań)")
plt.ylabel("prawdopodobieństwo")
plt.title("--- 5 ---")
count, bins, ignored = plt.hist(session_lengths, len(set(session_lengths)), alpha=0.5, density=True)
plt.show()

# --- 6 ---
session_hours = []
for i in computers.values():
    for s in i.sessions:
        session_hours.append(s.startTime//3600)

plt.xlabel("godzina rozpoczęcia sesji")
plt.ylabel("prawdopodobieństwo")
plt.title("--- 6 ---")
count, bins, ignored = plt.hist(session_hours, len(set(session_hours)), alpha=0.5, density=True)
plt.show()

# --- 7 ---
entry_points = []
for i in computers.values():
    for s in i.sessions:
        entry_points.append(s.requests[0].page)

plt.xlabel("strona rozpoczynająca sesje")
plt.ylabel("liczba wytąpień")
plt.title("--- 7 ---")
count, bins, ignored = plt.hist(entry_points, len(PAGES), alpha=0.5)
plt.show()

# --- 8 ---
departure_points = []
for i in computers.values():
    for s in i.sessions:
        departure_points.append(s.requests[-1].page)

plt.xlabel("strona kończąca sesje")
plt.ylabel("liczba wytąpień")
plt.title("--- 8 ---")
plt.xticks(rotation = 'vertical')
count, bins, ignored = plt.hist(departure_points, len(PAGES), alpha=0.5)
plt.show()

# --- 9 ---
times = {}
for i in PAGES:
    times[i] = []
    
for i in computers.values():
    for s in i.sessions:
        for ir,r in enumerate(s.requests):
            # If its the last page visit of session append max session time
            if ir == len(s.requests)-1:
                times[r.page].append(900)
            
            else:
                times[r.page].append(s.requests[ir+1].time - r.time)
                
for i in times:
    times[i] = np.mean(times[i])
    
plt.xlabel("strony")
plt.ylabel("średni czas spędzony na stronie(min)")
plt.title("--- 9 ---")
plt.xticks(rotation = 'vertical')
xplot = []
yplot = []
for page, time in times.items():
    xplot.append(page)
    yplot.append(time/60)
plt.bar(xplot, yplot, alpha=0.5)
plt.show()

# --- 10 ---
print("10.")
full_sessions = []
for i in computers.values():
    for s in i.sessions:
        full_sessions.append(s)
        
full_sessions.sort(key=lambda s: s.startTime)
for i in range(10):
    curr_sess = full_sessions[i]
    for r in curr_sess.requests:
        print(r.page, "-> ", end="")
    print("END")
```
